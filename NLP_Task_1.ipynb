```python
'''
1. What is the difference between "Love" and "love" in NLP?

In Natural Language Processing (NLP), "Love" and "love" are treated as different words if text preprocessing like lowercasing is not applied.
This is because NLP models are case-sensitive by default.

For example:

"Love" may appear at the beginning of a sentence or as a proper noun.
"love" is generally used as a verb or common noun.

If we do not normalize the text (convert to lowercase), the model may treat them as two separate tokens, which increases vocabulary size unnecessarily.
Therefore, we usually convert all text to lowercase to maintain consistency.



2. What happens if stopwords are not removed?

If stopwords (like is, the, at, on, in, and) are not removed:
-The dataset may contain many unnecessary words.
-The model may focus on unimportant words instead of meaningful words.
-Processing time and memory usage may increase.
-Model accuracy may decrease in some tasks like text classification.
However, in some cases (like sentiment analysis), stopwords may be important.


3. Provide two real-world scenarios where removing stopwords can be harmful.
Scenario 1: Sentiment Analysis
Example:
"I do not like this product"
If we remove stopwords, it becomes:
"I like product"
This changes the meaning completely because "not" is important for understanding sentiment.

Scenario 2: Question Answering Systems
Example:
"What is the best way to learn Python?"
If stopwords are removed:
"best way learn Python"
The grammatical meaning may be affected, which can reduce accuracy in question answering systems.


4. Explain the difference between stemming and lemmatization.

Stemming:
Stemming is a process of reducing words to their root form by removing suffixes, often without considering the meaning of the word.
The output may not be a valid word.
Example:
playing → play
studies → studi
running → run
It is faster but less accurate.

Lemmatization:
Lemmatization reduces words to their base form (lemma) using vocabulary and morphological analysis. The output is always a meaningful word.
Example:
playing → play
better → good
running → run
It is slower than stemming but more accurate.

'''
```

# Task 2:Build Advanced Preprocessing Function

```python
import nltk
import re
from collections import Counter

def preprocess_text(text):
    
    # Error handling (Task 7)
    if not text or not isinstance(text, str):
        return [], ""
    
    # Convert to lowercase
    text = text.lower()
    
    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)
    
    # Remove emails
    text = re.sub(r'\S+@\S+', '', text)
    
    # Remove numbers
    text = re.sub(r'\d+', '', text)
    
    # Handle repeated characters (loooove → love)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    
    # Remove emojis & special characters
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Remove very short tokens except exceptions
    exceptions = ['no','not']
    
    tokens = text.split()
    
    tokens = [word for word in tokens if len(word) > 2 or word in exceptions]
    
    cleaned_sentence = " ".join(tokens)
    
    return tokens, cleaned_sentence
```

# Task 3: Stress Testing

```python
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

all_tokens = []

for s in sentences:
    
    tokens, cleaned = preprocess_text(s)
    
    all_tokens.extend(tokens)
    
    print("Original Text:",s)
    print("Cleaned Tokens:",tokens)
    print("Cleaned Sentence:",cleaned)
    print("-"*50)
```

# Task 4: Token Analytics

```python
for s in sentences:
    
    tokens, cleaned = preprocess_text(s)
    
    if len(tokens)==0:
        avg_len = 0
    else:
        avg_len = sum(len(w) for w in tokens)/len(tokens)
    
    print("Sentence:",s)
    print("Total tokens:",len(tokens))
    print("Unique tokens:",len(set(tokens)))
    print("Average token length:",round(avg_len,2))
    print("-"*50)
```

# Task 6: Full Pipeline Function

```python
def full_pipeline(text_list):
    
    all_tokens = []
    clean_sentences = []
    
    for text in text_list:
        
        tokens, cleaned = preprocess_text(text)
        
        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)
    
    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
result = full_pipeline(sentences)

print(result)
```
